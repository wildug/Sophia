```python
adam = data["gpt2-small-adamW-10k - val/loss"]
sophia = data["gpt2-small-sophiag-10k - val/loss"]
steps = data["Step"]

fig, ax = plt.subplots(1)

ax.plot(steps, adam, label="AdamW")
ax.plot(steps, sophia, label="SophiaG")
ax.set_xlabel("Steps")
ax.set_ylabel("Validation Loss")
ax.legend()
plt.savefig("results/steps_adam_sophia.pdf")
```
